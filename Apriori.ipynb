```python
!pip install apyori
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header= None)
```

```python
no_orders = len(dataset)
len(dataset)
```

```python
no_items = len(list(dataset))
len(list(dataset))
```

```python
transactions = []
for i in range(0, no_orders):
  transactions.append([str(dataset.values[i, j]) for j in range(0, no_items)])
```

```python
rules = apriori(transactions= transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 3)
```

```python
results = list(rules)
results
```

```python
results = pd.DataFrame(results)
```

```python
comb = results['items']
comb = pd.DataFrame(comb)
comb
```
